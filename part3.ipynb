```python
sns.heatmap(df_6.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_all = df_6[df_6.columns[3:]]
df_all.dtypes
```

```python
def custom_melt(x,y): # GDP , Population
    df_2008 = df_all[[ f'{x}[2008]' , f'{y}[2008]' ]].copy()
    df_2018 = df_all[[ f'{x}[2018]' , f'{y}[2018]' ]].copy()
    df_2008['Years'] = 2008
    df_2018['Years'] = 2018
    df_2018 = df_2018.rename(columns={f'{x}[2018]':f'{x}' , f'{y}[2018]': f'{y}'})
    df_2008 = df_2008.rename(columns={f'{x}[2008]':f'{x}' , f'{y}[2008]': f'{y}'})
    df_out = df_2008.append(df_2018)
    return df_out
```

```python
cols = ["GDP" , "Population" , "Unemployment" , "Birth_rate" , "Population(15-64)" , "Population(>65)"]
for i in range(3):
    for j in range(3,6):
        df = custom_melt(cols[i],cols[j])
        plt.figure(figsize=(5,5))
        corr = df[cols[i]].corr(df[cols[j]])
        corr = round(corr,2)
        ax = sns.scatterplot(data=df , x=df[f'{cols[i]}']  , y=df[f'{cols[j]}'] , hue='Years')
        ax.set_title(f' {cols[i]} and {cols[j]} , Correlation = {corr}')
```
